```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python

data  ={
    'age': [25, 27, 29, np.nan, 32, 35, 30, 24, np.nan, 28],
    'salary': [50000, 54000, 58000, 62000, np.nan, 66000, 70000, 74000, 78000, 82000],
    'gender': ['M', 'F', 'F', 'M', np.nan, 'F', 'M', 'M', 'F', 'M'],
    'purchased': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}
```

```python
df=pd.DataFrame(data)
```

```python
df.head(10)
```

#Handeling missing data


```python
df.isnull().sum()
```

```python
df['age'].mean()
```

```python
#using simple imputer we can assign trhe mean,meadian,mode
numerical=SimpleImputer(strategy='mean')
categorical_imputer= SimpleImputer(strategy='most_frequent')
```

```python
df['age'] = numerical.fit_transform(df[['age']])
df['salary'] = numerical.fit_transform(df[['salary']])
df['gender'] = categorical_imputer.fit_transform(df[['gender']]).ravel()
```

```python
df.head(10)
```

#Hot encoding

```python
label_Encoder=LabelEncoder()
df['gender']=label_Encoder.fit_transform(df['gender'])
df['purchased']=label_Encoder.fit_transform(df['purchased'])
```

```python
df.head()
```

3.# Feature Scaling

```python
scaler=StandardScaler()
df[['age','salary']]=scaler.fit_transform(df[['age','salary']])
```

#Normilazation

```python
min_maxscaler=MinMaxScaler()
df[['age','salary']]=min_maxscaler.fit_transform(df[['age','salary']])
```

```python
# 4. Feature Transformation
# Log Transformation (if applicable)
# df['salary'] = np.log(df['salary'])

# 5. Dimensionality Reduction
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df[['age', 'salary']])
df['pca1'] = pca_components[:, 0]
df['pca2'] = pca_components[:, 1]
```

```python
# 7. Data Sampling
# Splitting data into train and test sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

print("Processed DataFrame:")
print(df)

print("\nTraining DataFrame:")
print(train_df)

print("\nTesting DataFrame:")
print(test_df)
```
